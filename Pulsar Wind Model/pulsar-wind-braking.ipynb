### Introduction

A Pulsar is a rotating compact star such as neutron stars emitting beams of electromagnetic radiation from its poles. This radiation can be observed when the beam is passing through Earth. Due to the rotation, this observation will appear like pulses, and therefore why this object is called Pulsar. There are many pulsars that have been identified, one of them is the Crab Pulsar (PSR B0531+21). Crab Pulsar is a relatively young pulsar discovered in 1968. This pulsar is related to a supernova SN 1054 that was observed in 1054.


The Crab Pulsar is observed to has a spin period of $P = 33.3924123 \;ms$ and a period derivative of $\dot{P} = 4.20972 \times 10^{-13}\;s/s$ and a characteristic age of $\tau_c = P/2\dot{P} \approx 1260\;years$ (Manchester et al. 2005). Its braking index was found to be $n = 2.51 \pm 0.01$ (Lyne et al. 1993).


The spin-down behavior of a pulsar is described by the following formula (Shapiro & Teukolsky 1983):

$$
\dot{\Omega} = -K\Omega^{n},
$$

where $K$ is a constant and $n$ is the braking index. Given this formula, the braking index can be defined as (Shapiro & Teukolsky 1983):

$$
n = -\frac{\Omega\ddot{\Omega}}{\dot{\Omega}^2}
$$

where $\Omega$ is the spin angular velocity, $\dot{\Omega}$, and $\ddot{\Omega}$ are the first, second, and third derivative.

The spinning down of pulsar can be described as a rotating dipole subject to a magnetic dipole radiation in vacuum. The magnetic moment has both a parallel and perpendicular component. For the spin-down caused by magnetic dipole radiation, the perpendicular component ($\mu_{\bot} = \mu\sin\alpha$) is responsible for it. The spin-down and alignment torque is given as (Michel & Goldwire 1970; Davis & Goldstein 1970):

$$
\begin{aligned}
\frac{d\Omega}{dt} &= -\frac{2\mu^{2}\Omega^{3}}{3Ic^{3}}\sin^{2} \alpha \\
\frac{d\alpha}{dt} &= -\frac{2\mu^{2}\Omega^{2}}{3Ic^{3}}\sin\alpha\cos\alpha
\end{aligned}
$$

where $\Omega = 2\pi/P$ is the angular speed of the pulsar and $\mu$ is the magnetic dipole moment of the pulsar, $\alpha$ is the inclination angle between rotational and magnetic axis, and $c$ is the speed of light. Using the formula for braking index, the braking index for this model is found to be (Michel & Goldwire 1970; Davis & Goldstein 1970):

$$
n = 3 + 2 \cot^{2}\alpha
$$

This braking index will always be greater than 3 (Eksi et al. 2016). Given that Crab Pulsar braking index was measured to be lower than three, that means there are other factor that are affecting the braking mechanism of crab pulsar.

There are many models that had been proposed to explain Crab Pulsar braking mechanism, one such model is the pulsar wind model (Xu & Qiao 2001; Wu et al. 2003; Contopoulos & Spitkovsky 2006; Yue et al. 2007).

### Evolution of the Crab Pulsar in the Pulsar Wind Model

In the pulsar wind model, the spin-down and the particle acceleration is taken into account (Xu & Qiao 2001). The pulsar wind are composed of accelerated particles and the parallel component ($\mu_{\parallel} = \mu\cos\alpha$) of the magnetic dipole moment is responsible for the particle acceleration (Ruderman & Sutherland 1975). Acceleration gaps are formed above the polar gap, primary particles are generated and accelerated in these gaps (Kou & Tong 2015). The energy that these particles taken is dependent on the acceleration potential drop. The rotational energy loss rate is (Xu & Qiao 2001):

$$
\dot{E_{p}} = 2\pi r^{2}_{p}c\rho_{e}\Delta{\phi}
$$

where $r_{p} = R(R\Omega/c)^{1/2}$ polar gap radius, $\rho_{e} = \kappa\rho_{GJ}$ is the primary particle density and $\rho_{GJ} = \Omega B/(2\pi c)$ is the Goldreich-Julian charge density (Goldreich & Julian 1969), $\Delta{\phi}$ is the corresponding acceleration potential in the gap, and $\kappa$ is a coefficient for the primary particle density. The maximum acceleration potential for a rotating dipole is $\Delta{\Phi} = \mu\Omega^{2}/c^{2}$ (Ruderman & Sutherland 1975). Then the rotational energy loss rate becomes:

$$
\begin{aligned}
\dot{E_{p}} &= \frac{2\mu^{2}\Omega^{4}}{3c^{3}}3\kappa\frac{\Delta{\phi}}{\Delta{\Phi}}\cos^{2}\alpha \\
-I\Omega\dot{\Omega} &= \frac{2\mu^{2}\Omega^{4}}{3c^{3}}3\kappa\frac{\Delta{\phi}}{\Delta{\Phi}}\cos^{2}\alpha
\end{aligned}
$$

Combining the magnetic dipole radiation equation and the equation above for particle wind gives:

$$
\begin{aligned}
-I\Omega\dot{\Omega} &= \frac{2\mu^{2}\Omega^{4}}{3c^{3}}\left(\sin^{2}\alpha + 3\kappa\frac{\Delta{\phi}}{\Delta{\Phi}}\cos^{2}\alpha\right) \\
&= \frac{2\mu^{2}\Omega^{4}}{3c^{3}}\eta
\end{aligned}
$$

The first and second term in $\eta$ are responsible for the braking by magnetic dipole radiation and particle wind respectively. If the acceleration potential is 0, there are no accelerated particles in the gap, and the pulsar is just braked down by the magnetic dipole radiation (Kou & Tong 2015). The acceleration potential is model dependent (Xu & Qiao 2001; Wu et al. 2003). The model being tested in this text is the pulsar wind model of Xu & Qiao (2001), with acceleration model of Vacuum Gap (VG) in which Vacuum gap forms above polar cap with enough binding energy where relativistic primary particles emit $\gamma$-rays via curvature radiation (CR) in the gap (Ruderman & Sutherland 1975).

Spin down equation in wind braking model:
$$
I\frac{d\Omega}{dt} \equiv -\frac{2\mu^2\Omega^3}{3c^3}\eta
$$
where (using the vacuum gap VG(CR) model),
$$
\eta \equiv \sin^2\alpha + 4.96\times 10^2\kappa B^{-8/7}_{12} \Omega^{-15/7} \cos^2\alpha
$$
where $B_{12}$ is the magnetic field in units of $10^{12}\;{\rm G}$, and $\kappa$ is the multiplier for  Goldreich-Julian charge density to find the primary particle density. The alignment torque for this model is the same as with the magnetic dipole radiation braking.

Pulsar death must be considered in modeling the long-term rotational evolution of the pulsar. When the pulsar death, primary particles generation ceases. As the pulsar is spinning down, spin angular velocity $\Omega$ of the pulsar drops and the maximum potential of the pulsar ($\Delta{\Phi}$) drops as well, and once it reaches the point where it could not accelerate particles anymore ($\Delta{\phi} > \Delta{\Phi})$, the pulsar death (Kou & Tong 2015). With no particle being accelerated in the magnetosphere, the effect of particle wind will also vanish, leaving the pulsar to be braked only by magnetic dipole radiation. This will modify $\eta$ into a piecewise function (Kou & Tong 2015):

$$
\eta \equiv \left\{
\begin{array}{ll}
    \sin^2\alpha + 4.96\times 10^2\kappa \left(1 - \frac{\Omega_{death}}{\Omega} \right) B^{-8/7}_{12} \Omega^{-15/7} \cos^2\alpha,\, if\; \Omega > \Omega_{death} \\
    \sin^2\alpha,\, if\; \Omega < \Omega_{death} \\
\end{array}
\right.
$$

The death period is defined as (Contopoulos & Spitkovsky 2006; Tong & Xu 2012):

$$
\Omega_{death} = \frac{2\pi}{P_{death}}
$$

where

$$
P_{death} = 2.8\left( \frac{B}{10^{12} \,G} \right)^{1/2} \left( \frac{V_{gap}}{10^{12} \,V} \right)^{-1/2} \,s.
$$

and $V_{gap} \equiv 10^{13} V$ (Contopoulos & Spitkovsky 2006).

The following code is used to test this model

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
from scipy.integrate import odeint
from matplotlib import rc

#comment out the following line for faster graph rendering and set usetex=False
rc('font', **{'family': 'sans-serif', 'sans-serif': ['Computer Modern']})
rc('text', usetex=True)
```

```python
I = 1e45     #g cm^2
R = 1e6     #cm6
B = 8.1   #G12
c = 2.998e10 #cm s^-1
k = 1000
year = 3.154e7  # s
Vgap = 1e13

mu = 0.5*(B*1e12)*(R**3)   #no need for the paranthesis
A = 2*mu**2 / (3*I*c**3)        #numerical constant
P_death = 2.8*np.sqrt(B)*((Vgap/1e12)**(-1/2))
omega_death = 2.0*np.pi/P_death
```

```python
#system of diff eqs. dy/dt = f(y,t)
def f(y, t):
    omega = y[0]
    alpha = y[1]
    
    C = 4.96*100*k*(B**(-8/7))*(np.cos(alpha*np.pi/180.0)**2)
    #vacuum gap model test
    if omega > omega_death:
        eta = np.sin(alpha*np.pi/180.0)**2 + C*(1-(omega_death/omega))*(omega**(-15/7))
    elif omega < omega_death:
        eta = np.sin(alpha*np.pi/180.0)**2
    
    #alignment torque same case with magnetosphere
    alignment_torque = np.sin(alpha*np.pi/180.0) * np.cos(alpha*np.pi/180.0)
    
    
    f0 = -(A * omega**3 * eta)
    f1 = -(A * omega**2 * alignment_torque)
    return [f0,f1]
```

Parameters for the initial conditions are taken from Kou & Tong (2015):

| $\kappa$    | $\alpha_{0}$  | $B_{12}$ |    $P_0$   |
| ----------- | ------------- | -------- | ---------- |
| $1000$      | $55^{\circ}$  |  $8.1$   | $18.8\;ms$ |

where $P_0$ is the initial period of the Crab Pulsar at birth and $\alpha_{0}$ is the initial inclination angle.

```python
#initial conditions
P0 = 0.0188  #s
omega0 = 2.0*np.pi/P0
alpha0 = 55  #degrees
y0 = [omega0, alpha0]
#time grid in year from year 1 to 30000
T = np.logspace(1.0,5.5,50000) / 10.   

#solve the DEs
t = T*year  #convert time unit from year to second for input
soln = odeint(f,y0,t)
omega = soln[:,0]
P = 2.0*np.pi/omega
alpha = soln[:, 1]
```

The braking index for this model is:

$$n = 3 + \frac{\Omega}{\eta}\frac{d\eta}{d\Omega}$$

The derivative term is:

$$
\begin{aligned}
\frac{d\eta}{d\Omega} &= \frac{d}{d\Omega}\left[ \sin^2\alpha + 4.96\times 10^2\kappa B^{-8/7}_{12} \Omega^{-15/7} \cos^2\alpha \right] \\
&= \sin\,(2\alpha)\frac{d\alpha}{d\Omega} + 4.96\times 10^2\kappa B^{-8/7}_{12} \frac{d}{d\Omega} \left[\Omega^{-15/7} \cos^2\alpha \right] \\
&= \sin\,(2\alpha)\frac{d\alpha}{d\Omega} + 4.96\times 10^2\kappa B^{-8/7}_{12} \left[-\frac{15}{7}\Omega^{-22/7}\cos^2\alpha - \Omega^{-15/7} \sin (2\alpha) \frac{d\alpha}{d\Omega}\right]
\end{aligned}
$$
where,
$$\frac{d\alpha}{d\Omega} = 0$$

So,
$$
\begin{aligned}
\frac{d\eta}{d\Omega} &= 4.96\times 10^2\kappa B^{-8/7}_{12} \left(-\frac{15}{7}\Omega^{-22/7}\cos^2\alpha \right)
\end{aligned}
$$

let $C = 4.96\times 10^2\kappa B^{-8/7}_{12}\cos^2\alpha$,
So the braking index is:

$$
n = 3+\frac{\Omega}{\eta} C \left(-\frac{15}{7}\Omega^{-22/7}\right)
$$

```python
C = 4.96*100*k*(B**(-8/7))*(np.cos(alpha0*np.pi/180.0)**2)  #numerical constant
eta = np.sin(alpha0*np.pi/180.0)**2 + C*(omega**(-15/7))
right_term = (C*omega/eta)*((-15/7)*(omega**(-22/7)))
n = 3 + right_term
```

```python
ftsz = 20
plt.rcParams['xtick.labelsize']=16
plt.rcParams['ytick.labelsize']=16
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(T,P, linewidth=3)
ax.plot(966,0.0333924123, 'or', markersize=5)  #Observed period in current year
arrow = mpatches.FancyArrowPatch((500,0.3), (955,0.034), 
                                 arrowstyle="-|>,head_width=0.05,head_length=0.1", 
                                 mutation_scale=100, fc='r', ec='r')
ax.add_patch(arrow)
ax.text(100, 0.32, 'Observed period in current year = 0.03339 s at 966 years old', 
        ha='center', va='center', color='r', fontsize=16)
ax.set_xscale("log")
ax.grid()
ax.set_xlabel('Time (yr)', fontsize=ftsz)
ax.set_ylabel('Period (s)', fontsize=ftsz)
ax.set_title('Period vs Time', fontsize=ftsz)
ax.set_xlim(1,3e4)
fig.tight_layout(pad=10.0)
plt.show()
```

The above figure shows the evolution of the period of the Crab Pulsar from birth to 30000 years of age in the VG(CR) model. The red arrow shows the current observed period of Crab Pulsar at 966 years old, which is $33.3924123\;ms$. As seen from the figure, the period grows slowly at the start and then rapidly increases in its later age. 

This indicates that the spinning down of Crab Pulsar is braked mainly by magnetic dipole radiation in its early age, and then mainly by particle wind later. From the calculation, this model predicts a period of $33.59653\;ms$ when Crab Pulsar is at its current age (966 years old).

The calculated inclination angle of this model is within the observation range.

The above figure plots the evolution of the braking index as a function of spin period. As the period increases, the braking index decreases from 3 to 1. The Crab pulsar is initially braked by magnetic dipole when the braking index is close to 3, and then the effect of particle wind starts to dominate when the braking index is approaching 1 (Kou & Tong 2015). The period at transition point ($n = 2$) is found to be $P_t = 54\;ms$. Crab Pulsar age is at $2689$ years old on the transition point.

To obtain the period derivative, starting from the spin-down equation and given that $\Omega = 2\pi/P$,

$$
\begin{aligned}
\frac{d}{dt}\Omega &= -\frac{2\mu^{2}\Omega^{3}}{3Ic^{3}}\eta \\
\frac{d}{dt} \left(\frac{2\pi}{P}\right) &= -\frac{2\mu^{2}(2\pi)^{3}}{3Ic^{3}P^{3}}\eta \\
-\frac{2\pi}{P^{2}}\dot{P} &= -\frac{2\mu^{2}(2\pi)^{3}}{3Ic^{3}P^{3}}\eta
\end{aligned}
$$

let $A = 2\mu^{2}/3Ic^{3}$, therefore,

$$
\dot{P} = A\frac{(2\pi)^{2}}{P}\eta
$$

```python
C = 4.96*100*k*(B**(-8/7))*(np.cos(alpha*np.pi/180.0)**2)  #numerical constant
eta = np.sin(alpha*np.pi/180.0)**2 + C*(omega**(-15/7))
eta_death = np.sin(alpha*np.pi/180.0)**2 + C*(1-(omega_death/omega))*(omega**(-15/7))
P_derivative = (A*(2*np.pi)**2)*eta_death/P
P_derivative_no_death = (A*(2*np.pi)**2)*eta/P
```

```python
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(T,P_derivative, linewidth=3, label='Accounting for pulsar death')
ax.plot(T,P_derivative_no_death, '-.', linewidth=3, label='Not accounting for pulsar death')
ax.plot(np.full(T.shape,T[T>=2689][0]),np.linspace(3e-13,6.5e-13,np.size(T)), '--g', 
        linewidth=3, label='Transition Point')
ax.plot(966,4.20972e-13, 'or', markersize=6)
arrow = mpatches.FancyArrowPatch((100,4.2e-13), (900,4.2e-13), 
                                 arrowstyle="-|>,head_width=0.05,head_length=0.1", 
                                 mutation_scale=100, fc='r', ec='r')
ax.add_patch(arrow)
ax.text(50, 4.25e-13, r'Observed Period Derivative of $4.21 \times 10^{-13}$ s/s at 966 years old', 
        ha='center', va='center', color='r', fontsize=14)
ax.set_xscale("log")
ax.grid()
ax.set_xlabel('Age (yr)', fontsize=ftsz)
ax.set_ylabel('Period Derivative (s/s)', fontsize=ftsz)
ax.set_title('Period derivative vs Time', fontsize=ftsz)
ax.set_xlim(1,3e4)
ax.set_ylim(3.5e-13,6.5e-13)
ax.legend(loc='lower left', fontsize=18)
fig.tight_layout(pad=10.0)
plt.show()
```

The above figure shows the evolution of the period derivative of Crab Pulsar with age. The period derivative decreases steadily up until the transition point when it starts increasing rapidly, indicating that after the transition point, the particle wind is dominating the braking of Crab Pulsar compared to the early age when it was mainly braked by magnetic dipole radiation (Kou & Tong 2015). From the calculation, this model predicts a period derivative of $\dot{P} = 4.14 \times 10^{-13} s/s$

```python
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(P,P_derivative, '--',linewidth=3, label='Accounting for pulsar death')
ax.plot(P,P_derivative_no_death, '--',linewidth=3, label='Not accounting for pulsar death')
ax.plot(P[0],P_derivative[0], 'or', markersize=8, label='Crab Pulsar at 1 Year Old')
ax.plot(P[T>=966][0],P_derivative[T>=966][0], 'ok', markersize=8, label='Crab Pulsar at 966 Years Old')
ax.plot(P[T>=2689][0],P_derivative[T>=2689][0], 'oy', markersize=8, label='Crab Pulsar at 2689 Years Old')
ax.plot(P[T>=10000][0],P_derivative[T>=10000][0], 'om', markersize=8, label='Crab Pulsar at 10000 Years Old')
ax.grid()
ax.set_xlabel('Period (s)', fontsize=ftsz)
ax.set_ylabel('Period Derivative (s/s)', fontsize=ftsz)
ax.set_title(r'Evolution of the Crab pulsar in the $P-\dot{P}$ diagram', fontsize=ftsz)
ax.legend(loc='lower right', fontsize=18)
fig.tight_layout(pad=10.0)
plt.show()
```

The above figure shows the evolution of the Crab pulsar in the $P-\dot{P}$ diagram. Points chosen in the graph are the age of crab pulsar starting from birth, to its current age, and the age at transition point, and finally at 10000 years old. As seen in the figure, the curve goes down at the start up to the transition point, and then evolves up sharply. The effect of pulsar death can also be seen in the figure. When the death of pulsar is not taken into account, the curve continually evolves in the same trend after the transition point. Following this trend will mean that the pulsar would not die and might end up as a magnetar in the long run (Kou & Tong 2015).

```python
fig, ax = plt.subplots(figsize=(16,16))
ax.loglog(radio_pulsar_data['P0'].values, radio_pulsar_data['P1_I'].values, '*r', label='Radio Pulsars')
ax.loglog(magnetars_data['P0'].values, magnetars_data['P1_I'].values, '^m', label='Magnetars')
ax.loglog(rrat_data['P0'].values, rrat_data['P1_I'].values, 'sg', label='RRAT')
ax.loglog(xins_data['P0'].values, xins_data['P1_I'].values, 'oc', label='XINS')
ax.loglog(cco_data['P0'].values, cco_data['P1'].values, 'dk', label='CCO')

log_grid = np.logspace(-3.0, np.log10(20), num=50, endpoint=True, base=10.0)

# Death Line
death_line = (log_grid**3) / (1.6e8**2)
ax.loglog(log_grid, death_line, linestyle='dashdot', color = '0.3')
ax.text(log_grid[log_grid >= 1e-2][0], death_line[death_line >= 1e-24][0], 
        r'Death Line', ha='center', rotation=38, fontsize=20)


# Magnetic Fields
B_scale = [1e8, 1e10, 1e12, 1e14]
for scale in B_scale:
    magnetic_field = (scale/3.2e19)**2 / log_grid
    ax.loglog(log_grid, magnetic_field, '--k', color = '0.3')

#ax.text(10, 1e-25, r'$10^{8}$ G', rotation=-11, fontsize=16)
ax.text(10, 1.5e-21, r'$10^{10}$ G', rotation=-11, fontsize=16)
ax.text(6.5, 1.5e-17, r'$B_c = 10^{12}$ G', rotation=-11, fontsize=16)
ax.text(10, 1e-12, r'$10^{14}$ G', rotation=-11, fontsize=16)


# Characteristics Age
hubble_time_scale = [1e3,1e6,1e9]
for scale in hubble_time_scale:
    hubble_time = log_grid / (2*scale*365*24*60*60)
    ax.loglog(log_grid, hubble_time, 'k:', color = '0.3')

ax.text(1.1e-3, 3.2e-14, r'$\tau_{c} = 10^3$ yr', rotation=9, fontsize=16)
ax.text(1.1e-3, 3.2e-17, r'$\tau_{c} = 10^6$ yr', rotation=9, fontsize=16)
ax.text(1.1e-3, 3.2e-20, r'$\tau_{c} = 10^9$ yr', rotation=9, fontsize=16)
    
    
# Crab Pulsar evolution    
ax.plot(P,P_derivative, '-b',linewidth=3, 
        label='Rotational evolution of the Crab pulsar accounting for pulsar death')
ax.plot(P,P_derivative_no_death, '--b',linewidth=3, 
        label='Rotational evolution of the Crab pulsar not accounting pulsar death')
ax.plot(P[0],P_derivative[0], 'p', markersize=8, label='Crab Pulsar at 1 Year Old')
ax.plot(P[T>=966][0],P_derivative[T>=966][0], 'pb', markersize=8, label='Crab Pulsar at 966 Years Old')
ax.plot(P[T>=2689][0],P_derivative[T>=2689][0], 'py', markersize=8, label='Crab Pulsar at 2689 Years Old')
ax.plot(P[T>=10000][0],P_derivative[T>=10000][0], 'p', markersize=8, label='Crab Pulsar at 10000 Years Old')
ax.set_xlim(10**(-3), 20)
ax.set_ylim(10**(-27), 10**(-8))
ax.set_title(r'$P-\dot{P}$ diagram', fontsize=ftsz)
ax.set_xlabel('Period (s)', fontsize=ftsz)
ax.set_ylabel('Period Derivative (s/s)', fontsize=ftsz)
ax.legend(loc='lower right', fontsize=14)
```

The figure above shows the Crab Pulsar on the $P-\dot{P}$ diagram. Pulsars with radio emission are denoted with a red star. Magnetars are denoted with magenta triangles. Rotating radio transient (RRAT) are shown with a green square. X-ray dim isolated neutron stars (XINS) are denoted by cyan circles. Central compact objects (CCO) are denoted by black diamond shape. Lastly, the rotational evolution of the Crab Pulsar (PSR B0531+21) is denoted by the curve and the pentagon shapes are Crab Pulsar age from birth to 10000 years old. The dashed curve does not consider pulsar death. The data is taken from ATNF pulsar catalogue (Manchester et al. 2005).

As seen from the figure and the figure before, when pulsar death is not considered, the pulsar might not die and end up  as a magnetar. With that, it will be difficult to explain the observations of the properties of the Crab pulsars that are significantly different from magnetars (Ng & Kaspi 2011), and that only a small portion of magnetars are radio emitters (Olausen & Kaspi 2014). Therefore, the effect of pulsar death must be considered in modelling the long-term rotational evolution.

The effect of pulsar death is negligible in the early age of Crab Pulsar as they evolve with the same trend in the figure. The trend starts changing after the transition point in which the effect of pulsar death start to show as it approaches the death period.
